**Axis Functions**

1) axes --> adds axis to the figure

2) text --> add text to the axes

3) title --> set the title of the current axes

4) xlabel --> set the x-axis label

5) xlim --> set the limit of the x axis

6) xscale --> set the scaling of the x axis

7) xticks --> get or set the x-limit of the current tick location /  labels

8) ylabel --> set the y-axis label

9) ylim --> set the limit of the y axis

10) yscale --> set the scaling of the y axis

11) yticks --> get or set the ylimit of the current tick location /  labels

**Figure**

1) A figure object is the outermost container for matplotlib plot

2) the figure object consist of multiple axis objects.

3) The figure is the final graphic that contains one or more plots

**Histogram**

```python
data = np.random.randn(1000)
```

```python
figure = plt.figure(figsize=(10,9),linewidth=9,edgecolor='black')
ax = plt.gca()
ax.set_facecolor('red')
plt.hist(data,edgecolor='black',linewidth=2,color='yellow')
plt.xlabel('X Axis')
plt.ylabel('Count')
plt.title('MY HISTOGRAM',fontsize=25)
```

## bar plot

```python
programming_lang = ['Python','Java','C','Go','Ruby']
no_of_students = [50,33,27,35,28]
```

```python
c = ['red','black','green','orange','blue']
plt.bar(programming_lang,no_of_students,width=0.5,color=c)
plt.xlabel('Programming Language')
plt.ylabel('No of Students')
```

```python
c = ['red','black','green','orange','blue']
plt.barh(programming_lang,no_of_students,color=c)
plt.xlabel('No of Studentse')
plt.ylabel('Programming Language')
```
