```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
help(sns.load_dataset)
```

```python
tips_data = sns.load_dataset('tips')
```

```python
tips_data
```

```python
sns.set_style(style='whitegrid')
```

```python
sns.lineplot(x='size',y='total_bill',data=tips_data)
```

```python
sns.scatterplot(x='size',y='total_bill',data=tips_data)
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips_data)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=tips_data)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=tips_data)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='time',data=tips_data)
```

```python
my_marker = {'Male':"P",'Female':"X"}
colors = ['#FF0000','#89CFF0']
sns.set_palette(sns.color_palette(colors))
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=tips_data,style='sex',markers=my_marker,legend=False)
```
